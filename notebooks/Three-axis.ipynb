```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xscale("log")
ax.set_yscale("log")

# Define ranges for axes
x_vals = np.logspace(1, 6, 100)
y_vals = np.logspace(-5, -1, 100)

# Plot the NISQ box
ax.fill_betweenx([10**-3, 10**-2], 10, 1000, color="lightblue", alpha=0.5, label="NISQ")

# Plot labels and markers
#ax.text(15, 3e-2, "Noisy Intermediate Scale\nQuantum Computing", verticalalignment='center', fontsize=12)
#ax.plot(30, 10**-2.5, 'ko')  # Today's marker
#ax.text(30, 10**-2.3, "Today", verticalalignment='center', fontsize=10)

ax.plot(100, 10**-3.5, 'ko')  # Next 5-10 years marker
ax.text(100, 10**-3.2, "Next 5-10 years", verticalalignment='center', fontsize=10)

# Current error correction threshold line
ax.axhline(y=10**-3, color="black", linestyle="-", label="Current Error Correction Threshold")

# Fault-Tolerant Quantum Computing (FTQC) area
ax.fill_betweenx(y_vals, 1000, 10**6, where=(y_vals < 10**-3), color="lightblue", alpha=0.3, label="Fault Tolerant Quantum Computing - FTQC")

# Labels for the FTQC area
ax.text(3000, 10**-4, "Fault Tolerant Quantum\nComputing - FTQC", fontsize=12, verticalalignment='center')

# Add arrows
#ax.annotate("", xy=(10**3.5, 10**-4.5), xytext=(10**1.5, 10**-1.5),
#            arrowprops=dict(arrowstyle="->", lw=1.5))
#ax.text(10**3.8, 10**-4.8, "Minimize Error Rates", rotation=270, verticalalignment="bottom", fontsize=12)

#ax.annotate("", xy=(10**5.8, 10**-3.8), xytext=(10**1.3, 10**-3.8),
#            arrowprops=dict(arrowstyle="->", lw=1.5))
#ax.text(10**4, 10**-3.7, "Increase qubits", horizontalalignment="center", fontsize=12)

# Set axis labels
ax.set_xlabel("Number of Physical Qubits", fontsize=14)
ax.set_ylabel("Physical Error Rates", fontsize=14)

# Customize ticks
ax.set_xticks([10, 100, 1000, 10000, 100000, 1000000])
ax.set_xticklabels(['10', '100', '1,000', '10,000', '100,000', '1 Million'])
ax.set_yticks([10**-5, 10**-4, 10**-3, 10**-2, 10**-1])
ax.set_yticklabels(['$10^{-5}$', '$10^{-4}$', '$10^{-3}$', '$10^{-2}$', '$10^{-1}$'])

# Display plot
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.legend()
plt.show()
```
